```python
import numpy as np 
import pandas as pd 
train_set = pd.read_csv('adult.data', header = None)
val_set = pd.read_csv('adult.test' , skiprows = 1, header = None) 
col_labels = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',  'marital_status', 'occupation','relationship', 'race', 'sex', 'capital_gain',  'capital_loss', 'hours_per_week', 'native_country', 'wage_class'] 
train_set.columns = col_labels 
val_set.columns = col_labels 
```

```python
df = pd.concat([train_set,val_set])

df.head()
```

```python
df.replace(' ?',np.nan,inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(df['wage_class'])
```

```python
df = df.replace({' <=50K':0,' >50K':1,' <=50K.':0,' >50K.':1})
```

```python
sns.countplot(df['wage_class'])
```

```python
df= df.replace(' Never-worked',' Without-pay')
```

```python
df['workclass'].value_counts()
```

```python
df['workclass'].fillna('0',inplace=True)
```

```python
plt.figure(figsize=(20,8))
sns.countplot(df.workclass)
```

```python
df['education'].value_counts()
```

```python
def school(x):
    if x in [' 1st-4th', ' 5th-6th', ' 7th-8th', ' 9th', ' 10th', ' 11th', ' 12th']:
        return 'School'
    else:
        return x
df['education'] = df['education'].apply(school)
```

```python
plt.figure(figsize=(20,8))

sns.barplot(x='education',y='wage_class',data=df)
```

```python
df['occupation'].unique()
```

```python
# No info is present 
df['occupation'].fillna('0',inplace=True)
# As there are many outliers in armed forces
df['occupation'].replace(' Armed-Forces','0',inplace=True)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='occupation',y='wage_class',data=df)
plt.xticks(rotation=30)
```

```python
def local(country):
    if country in [' United-States',' Canada']:
        return 'North_America'
    elif country in [' Puerto-Rico',' El-Salvador',' Cuba',' Jamaica',' Dominican-Republic',' Guatemala',' Haiti',' Nicaragua',' Trinadad&Tobago',' Honduras']:
        return 'Central_America' 
    elif country in [' Mexico',' Columbia',' Vietnam',' Peru',' Ecuador',' South',' Outlying-US(Guam-USVI-etc)']:
        return 'South_America'
    elif country in [' Germany',' England',' Italy',' Poland',' Portugal',' Greece',' Yugoslavia',' France',' Ireland',' Scotland',' Hungary',' Holand-Netherlands']:
        return 'EU'
    elif country in [' India',' Iran',' China',' Japan',' Thailand',' Hong',' Cambodia',' Laos',' Philippines',' Taiwan']:
        return 'Asia'
    else:
        return country
```

```python
df['native_country'] = df['native_country'].apply(local)
plt.figure(figsize=(20,10))

sns.barplot(x='native_country',y='wage_class',data=df)
```

```python
plt.figure(figsize=(10,12))
sns.heatmap(df.corr(),annot=True)
```

```python
X = df.drop(['wage_class','fnlwgt'],axis=1)
y = df['wage_class']
```

```python
X = pd.get_dummies(X)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.3,random_state=42)
```

```python
params = [{ 'learning_rate':[0.01,0.001],
                        'max_depth': [3,5,10],
                        'n_estimators':[10,50,100,200]
                    }
                   ]
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier
Xbc = XGBClassifier()
Gcv = GridSearchCV(Xbc,params,scoring='accuracy',cv=5,n_jobs=3,verbose=3)
Gcv.fit(x_train,y_train)
```

```python
XBC = XGBClassifier(learning_rate=0.01,max_depth=10,n_estimators=200)
XBC.fit(x_train,y_train)
```

```python
XBC.score(x_test,y_test)
y_pred = XBC.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

print(f'Accuracy Score:{accuracy_score(y_test,y_pred)}')
print('*'*50)
print(f'Confusion Matrix:{confusion_matrix(y_test,y_pred)}')
print('*'*50)
print(f'Classification Report: {classification_report(y_test,y_pred)}')
```
